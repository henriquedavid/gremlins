# Análise Gráfica do SLPool com o OS

```python
import pandas as pd
import matplotlib.pyplot as plt
fig, axes = plt.subplots(figsize=(10,7))
dados = pd.read_csv("qntAlocado.csv", sep=";")
dados.plot(x="Repetições", y=["SLPool", "SO"], color=["red", "green"], ax=axes)
plt.show()
```

   No gráfico acima é possível comparar a utilização entre o Gerenciador de Memória SLPool e o SO. É possível perceber que quanto maior a quantidade de dados que será trabalhando, assim como a quantidade de vezes que será necessário executar uma mesma ação é recomendável a utilização do SLPool. O SLPool é recomendado quando se utiliza grande quantidade de dados e ações, pois ele foi projetado para trabalhar melhor a memória de forma que ele tenha a melhor performance.
    
   Porém é possível notar no gráfico que é mais vantajoso a utilização do SO quando se trata com pequenas quantidades de dados, até mesmo porque é possível perceber que enquanto o SLPool apresentou um pico, o SO apresentou uma leve queda.
